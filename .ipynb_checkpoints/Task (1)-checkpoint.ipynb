```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

variable = 'most ads day'

# Specify the order of the days
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df_cat[variable] = pd.Categorical(df_cat[variable], categories=days_order, ordered=True)

# Create a figure
plt.figure(figsize=(12, 6))

# Count plot
plt.subplot(1, 2, 1)
sns.countplot(x=variable, data=df_cat, order=days_order)
plt.title(f'Count Plot {variable}')

# Pie chart
plt.subplot(1, 2, 2)
counts = df_cat[variable].value_counts().reindex(days_order)  # Reindex to ensure the correct order
plt.pie(counts, labels=counts.index, autopct='%0.1f%%')
plt.title(f'Pie Chart {variable}')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Вычисление эффекта размера (Коэн d)
def cohen_d(x, y):
    return (np.mean(x) - np.mean(y)) / np.sqrt(((np.std(x) ** 2) + (np.std(y) ** 2)) / 2)

effect_size = cohen_d(ad_conversions,psa_conversions)
print(f"Эффект размера (Коэн d): {effect_size:.4f}")
```

```python
import numpy as np
import statsmodels.api as sm

# Данные конверсий
conversions_ad = np.sum(df[df['test group'] == 'ad']['converted'])
total_ad = len(df[df['test group'] == 'ad'])
conversions_psa = np.sum(df[df['test group'] == 'psa']['converted'])
total_psa = len(df[df['test group'] == 'psa'])

# Проведение z-теста для пропорций
count = np.array([conversions_ad, conversions_psa])
nobs = np.array([total_ad, total_psa])

# Выполнение теста
z_stat, p_value = sm.stats.proportions_ztest(count, nobs)

print(f"Z-статистика: {z_stat:.4f}")
print(f"P-значение: {p_value:.4f}")

# Интерпретация результата
alpha = 0.05  # уровень значимости
if p_value < alpha:
    print("Разница в показателях конверсии статистически значима.")
else:
    print("Разница в показателях конверсии не является статистически значимой.")
```
